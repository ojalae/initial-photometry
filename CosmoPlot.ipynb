```python
import numpy as np
from matplotlib import pyplot as plt

import pandas as pd
from scipy.interpolate import interp1d

%matplotlib inline
```

```python
def PlotData(filename,columns):
    modelfile = filename
    redshift, distance = np.loadtxt(filename, usecols=columns, skiprows=0, unpack=True)
    plt.scatter(distance/1000.,redshift*3*10**5)
```

```python
PlotData('galaxydata.txt',(0,1))
plt.xlabel("Luminosity Distance [Gpc]")
plt.ylabel("cz (km/s)")
axes = plt.gca()
axes.set_ylim([0,900000])
axes.set_xlim([0,15])
```
